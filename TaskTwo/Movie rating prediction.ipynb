### Feature Engineering


This code performs Label Encoding on categorical columns in the movie DataFrame, such as 'Director', 'Actor 1', 'Actor 2', 'Actor 3', and 'Genre', to convert them into numerical values. The 'Director' and 'Actor' columns are encoded individually by applying fit_transform for each, while the 'Genre' column is first joined into a single string and then encoded. Finally, the resulting transformed data is stored in a new DataFrame called trans_data, with the 'Name' column dropped.

```python

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

trans_data = movie.drop(['Name'],axis=1)
# Transform Director columns
trans_data['Director'] = labelencoder.fit_transform(movie['Director'])

# Transform Actors Columns
trans_data['Actor 1'] = labelencoder.fit_transform(movie['Actor 1'])
trans_data['Actor 2'] = labelencoder.fit_transform(movie['Actor 2'])
trans_data['Actor 3'] = labelencoder.fit_transform(movie['Actor 3'])

trans_data['Genre'] = labelencoder.fit_transform(movie['Genre'].apply(lambda x: ', '.join(x)))

trans_data.head()
```

This code applies **Min-Max Scaling** to the `trans_data` DataFrame using `MinMaxScaler` from `sklearn.preprocessing`. It transforms the data to a range between 0 and 1 by scaling each feature in the dataset according to its minimum and maximum values. The scaled data is then converted into a new DataFrame `sc_df`, retaining the original column names from `trans_data`.

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sc_data = scaler.fit_transform(trans_data)
sc_df = pd.DataFrame(sc_data,columns=trans_data.columns)

sc_df.head()
```

```python
# correlation
corr_df = trans_data.corr(numeric_only=True)
corr_df['Rating'].sort_values(ascending=False)
```

calculates the correlation matrix of the trans_data DataFrame, considering only numeric columns, using the .corr() method. It then sorts the correlation values of the 'Rating' column in descending order to identify which variables are most strongly correlated with the rating.

Correlation is used to identify the strength and direction of the relationship between two or more variables, helping to understand how they influence each other.

## Model Building

```python
# Import modeling Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python

# put data except Rating data
X = trans_data.drop(['Rating'],axis=1)

# Put only Rating data
y = trans_data['Rating']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)
```

The above code splits the `trans_data` DataFrame into features (`X`) by dropping the 'Rating' column and the target variable (`y`) containing only the 'Rating' column, then it divides the data into training and testing sets using a 70-30 split. A **Linear Regression model** is then created and trained on the training data (`X_train`, `y_train`) to learn the relationship between the features and the target variable.

```python
X_test = np.array(X_test)
```

```python

y_pred = model.predict(X_test)
y_pred
```

```python


print('R2 score: ',r2_score(y_test,y_pred))
print('Mean squared error: ',mean_squared_error(y_test,y_pred))
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```

This code splits the `trans_data` DataFrame into features (`X`) and target (`y`), then divides them into training and testing sets. A **Linear Regression model** is trained on the training data, and predictions are made on the test set (`X_test`). The model's performance is evaluated using metrics like **RÂ² score**, **Mean Squared Error (MSE)**, and **Mean Absolute Error (MAE)**, and the actual test values (`y_test`) are printed for comparison.

Using this process, we evaluate the performance of the **Linear Regression model** that predicts the 'Rating' based on other features in the dataset. Specifically:

1. **RÂ² score** tells us how well the model explains the variance in the target variable ('Rating'). A higher RÂ² indicates a better fit.
2. **Mean Squared Error (MSE)** and **Mean Absolute Error (MAE)** provide insights into the model's accuracy by measuring the average difference between predicted and actual values. Lower values for MSE and MAE indicate better model performance.

In summary, these metrics help assess the model's predictive power and how closely its predictions match the actual ratings in the test data.
